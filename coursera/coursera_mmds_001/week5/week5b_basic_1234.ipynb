```python
%pylab inline
df = pd.DataFrame(points, columns=['x', 'y', 'color'])
plt.figure()
pd.DataFrame.plot(df, x='x', y='y', kind='scatter', c=df.color)
```

```python
centroids_green = df[df['color'] == 1]
points_yellow = df[df['color']==0]

def findCentroid(x,y,list_centroids):
    dist = list_centroids.apply(lambda row: np.linalg.norm(np.array((row['x'], row['y'])) - np.array((x,y))), axis=1)
    list_centroids['dist'] = dist
    result = list_centroids.ix[list_centroids['dist'].idxmin()]
    return str((result['x'], result['y']))
    #print list_centroids


points_yellow['closest_x'] = points_yellow.apply(lambda row: findCentroid(row['x'], row['y'], centroids_green), axis=1)
print points_yellow
print '--------------------'

new_centroids = {'x':[], 'y':[]}

for name, group in points_yellow.groupby('closest_x'):
    print 'name: ', name 
    print 'group: ', group
    xs = list(group['x'])
    xs.append(np.int64(eval(name)[0]))
    x_avg = np.mean(xs)
    ys = list(group['y'])
    ys.append(np.int64(eval(name)[1]))
    y_avg = np.mean(ys)
    new_centroids['x'].append(x_avg)
    new_centroids['y'].append(y_avg)

new_centroids = pd.DataFrame(new_centroids)

print '----------------------'
print 'new centroids: '
print new_centroids
print '----------------------------'
points_yellow['closest_x_new'] = points_yellow.apply(lambda row: findCentroid(row['x'], row['y'], new_centroids), axis=1)
print points_yellow

```
